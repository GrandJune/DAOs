```python
from Reality import Reality
from Individual import Individual
from Organization import Organization
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Search Iteration as the basis of the Performance Curve
m = 30
# s = 1
n = 50
p1 = 0.5
p2 = 0.5
loop = 100
results = []
for _ in range(100):
    reality = Reality(m=m)
    organization = Organization(m=m, n=n, p1=p1, p2=p2, reality=reality)
    organization.process(loop=loop)
    results.append(organization.performance_curve)
result_2 = []
for i in range(100):
    temp = [curve[i] for curve in results]
    result_2.append(sum(temp) / len(temp))
x = range(100)
plt.plot(x, result_2, "k-")
# plt.savefig("search.jpg")
plt.title('Performance Evolution')
plt.xlabel('Time')
plt.ylabel('Performance')
plt.show()
```

```python
# Search Iteration as the basis of the Performance Curve
m = 30
s = 1
n = 50
p1 = 0.5  # learning from code
p2 = 0.5  # learning from partial belief
loop = 100
task_size_list = [10, 20, 30]
results = []
for task_size in task_size_list:
    temp = []
    temp_3 = []
    for _ in range(100):
        reality = Reality(m=m, s=s)
        organization = Organization(m=m, s=s, n=n, p1=p1, p2=p2, reality=reality)
        organization.process(loop=loop, task_size=task_size)
        temp.append(organization.performance_curve)
    for i in range(100):
        temp_2 = [curve[i] for curve in temp]
        temp_3.append(sum(temp_2) / len(temp_2))
    results.append(temp_3)
x = range(100)
plt.plot(x, results[0], "k-", label="task_size=10")
plt.plot(x, results[1], "k--", label="task_size=20")
plt.plot(x, results[2], "k-.", label="task_size=30")
# plt.savefig("search.jpg")
plt.title('Performance Evolution')
plt.xlabel('Time')
plt.ylabel('Performance')
plt.legend()
plt.show()
```

```python
# Search Iteration as the basis of the Performance Curve
m = 30
s = 1
n = 50
p1_list = np.arange(0, 1.1, 0.1)  # learning from code
p2 = 0.5  # learning from partial belief
loop = 100
task_size_list = [10, 20, 30]
results = []

for task_size in task_size_list:
    temp_3 = []
    for p1 in p1_list:
        temp = []
        for _ in range(100):
            reality = Reality(m=m, s=s)
            organization = Organization(m=m, s=s, n=n, p1=p1, p2=p2, reality=reality)
            organization.process(loop=loop, task_size=task_size)
            temp.append(organization.performance_average)
        temp_3.append(sum(temp) / len(temp))
    results.append(temp_3)
x = p1_list
plt.plot(x, results[0], "k-", label="task_size=10")
plt.plot(x, results[1], "k--", label="task_size=20")
plt.plot(x, results[2], "k:", label="task_size=30")
# plt.savefig("search.jpg")
plt.title('Performance Across Task Size and P1')
plt.xlabel('P1-learning from code')
plt.ylabel('Performance')
plt.legend()
plt.show()
```
