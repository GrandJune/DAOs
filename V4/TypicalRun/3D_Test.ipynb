```python
import numpy as np
from scipy.signal import convolve2d
from scipy.interpolate import griddata
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# generate random data
data = np.random.rand(10, 10)

# create a 10x10 grid of coordinates
x = np.arange(0, 10, 1)
y = np.arange(0, 10, 1)
X, Y = np.meshgrid(x, y)

# plot the original surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, data)

# interpolate the data onto a denser grid
xi = np.linspace(0, 9, 20)
yi = np.linspace(0, 9, 20)
xi, yi = np.meshgrid(xi, yi)
zi = griddata((X.flatten(), Y.flatten()), data.flatten(), (xi, yi), method='cubic')

# plot the interpolated surface
ax.plot_surface(xi, yi, zi, cmap='viridis')

# calculate the ruggedness of the interpolated surface
laplacian = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])
ruggedness = np.abs(convolve2d(zi, laplacian, mode='valid')).sum()

# add the ruggedness value to the figure
ax.text2D(0.05, 0.95, f"Ruggedness: {ruggedness:.2f}", transform=ax.transAxes)
plt.savefig(data_folder + r"\Test_case_2.png", transparent=True, dpi=200, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from scipy.interpolate import griddata

# Generate some random data
data = np.random.rand(100)

# Reshape the data into a 10x10 grid
data = data.reshape((10, 10))

# Create a denser grid for interpolation
xi = np.linspace(0, 9, 100)
yi = np.linspace(0, 9, 100)
xi, yi = np.meshgrid(xi, yi)

# Interpolate the data onto the denser grid
zi = griddata((np.arange(10).repeat(10), np.tile(np.arange(10), 10)), data.flatten(), (xi, yi), method='cubic')
# Set up the figure and axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D surface
ax.plot_surface(xi, yi, zi, cmap='viridis')

# Set the axis labels
# ax.set_xlabel('X')
# ax.set_ylabel('Y')
ax.set_zlabel('Performance')
ax.set_xticks([])
ax.set_yticks([])
# Show the plot
plt.grid(visible=False)
plt.savefig(data_folder + r"\Typical_run.png", transparent=True, dpi=1000, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy import interpolate

# generate random data
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# create a denser grid
x_new = np.linspace(-5, 5, 50)
y_new = np.linspace(-5, 5, 50)
X_new, Y_new = np.meshgrid(x_new, y_new)

# interpolate the data onto the denser grid
Z_new = interpolate.griddata((X.ravel(), Y.ravel()), Z.ravel(), (X_new, Y_new), method='cubic')

# create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X_new, Y_new, Z_new)

# show plot
data_folder = "E:\data\dao-0310\V4_4\TypicalRun"
plt.savefig(data_folder + r"\Test_case_1.png", transparent=True, dpi=1000, bbox_inches='tight')
plt.show()
```
