```python
x = range(len(dao_performance))
group_size_list = [7, 14, 21, 28]
dao_convergence = [each[-1] for each in dao_performance]
hierarchy_convergence = [each[-1] for each in hierarchy_performance]
autonomy_convergence = [each[-1] for each in autonomy_performance]
fig, ax = plt.subplots()
ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)
plt.plot(group_size_list, dao_convergence, "-o", label="DAO", color=nature_orange)
plt.plot(group_size_list, hierarchy_convergence, "-s", label="Hierarchy", color=nature_green)
plt.plot(group_size_list, autonomy_convergence, "-v", label="Autonomy", color=nature_blue)
plt.xlabel('Group Size', fontweight='bold', fontsize=12)
plt.ylabel('Performance', fontweight='bold', fontsize=12)
plt.xticks(group_size_list)
plt.legend(frameon=False, ncol=1, fontsize=12)
plt.savefig(data_folder + r"\DHA_performance_across_group_size.png", transparent=True, dpi=300)
plt.show()
plt.clf()
```
