```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
legend_properties = {'weight':'bold'}
```

```python
data_folder = "/Volumes/T7/data/dao-0310-23/V4_4/Quadratic_Voting"

dao_performance_file = data_folder + "/dao_quadratic_performance"
consensus_performance_file = data_folder + "/dao_quadratic_consensus_performance"
dao_diversity_file = data_folder + "/dao_quadratic_diversity"
dao_variance_file = data_folder + "/dao_quadratic_variance"


with open(dao_performance_file, 'rb') as infile:
    dao_performance = pickle.load(infile)

with open(consensus_performance_file, 'rb') as infile:
    consensus_performance = pickle.load(infile)

with open(dao_diversity_file, 'rb') as infile:
    dao_diversity = pickle.load(infile)

with open(dao_variance_file, 'rb') as infile:
    dao_variance = pickle.load(infile)
```

```python
# Performance
x = range(len(dao_performance[0]))
fig, ax = plt.subplots()

ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)

plt.tight_layout()

plt.plot(x, dao_performance[0], "k-", label='$\\rho=0$')
plt.plot(x, dao_performance[1], "k:", label='$\\rho=1$')
plt.plot(x, dao_performance[2], "k--", label='$\\rho=2$')
plt.plot(x, dao_performance[3], "k-.", label='$\\rho=3$')

plt.xlabel('Time', fontweight='bold', fontsize=12)
plt.ylabel('Performance', fontweight='bold', fontsize=12)

plt.legend(frameon=False, ncol=1, fontsize=12, prop=legend_properties)

plt.savefig(data_folder + "/Asymmetry_performance.png",
            transparent=True,
            dpi=300)

plt.show()
plt.clf()
```

```python
# Linear Token-Weighted Voting
linear_data_folder = r"/Volumes/T7/data/dao-0310-23/V4_4/Asymmetry"
dao_performance_file = linear_data_folder + r"/dao_performance"
consensus_performance_file = linear_data_folder + r"/dao_consensus_performance"
dao_diversity_file = linear_data_folder + r"/dao_diversity"
dao_variance_file = linear_data_folder + r"/dao_variance"

with open(dao_performance_file, 'rb') as infile:
    linear_dao_performance = pickle.load(infile)
with open(dao_diversity_file, 'rb') as infile:
    linear_dao_diversity = pickle.load(infile)
with open(dao_variance_file, 'rb') as infile:
    linear_dao_variance = pickle.load(infile)

with open(consensus_performance_file, 'rb') as infile:
    linear_consensus_performance = pickle.load(infile)

```

```python
# Performance
x = range(len(dao_performance[0]))
fig, ax = plt.subplots()

ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)

plt.tight_layout()

plt.plot(x, linear_dao_performance[0], "k-", label='$\\rho=0$')
plt.plot(x, linear_dao_performance[1], "k:", label='$\\rho=1$')
plt.plot(x, linear_dao_performance[2], "k--", label='$\\rho=2$')
plt.plot(x, linear_dao_performance[3], "k-.", label='$\\rho=3$')

plt.xlabel('Time', fontweight='bold', fontsize=12)
plt.ylabel('Performance', fontweight='bold', fontsize=12)

plt.legend(frameon=False, ncol=1, fontsize=12, prop=legend_properties)

# plt.savefig(data_folder + "/Asymmetry_performance.png",
#             transparent=True,
#             dpi=300)

plt.show()
plt.clf()
```

```python
plot_asymmetry_levels = [1, 2, 3]

linear_performance_loss = [
    linear_dao_performance[0][-1] - linear_dao_performance[i][-1]
    for i in plot_asymmetry_levels
]

quadratic_performance_loss = [
    dao_performance[0][-1] - dao_performance[i][-1]
    for i in plot_asymmetry_levels
]

fig, ax = plt.subplots(figsize=(6, 4))

ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)

ax.plot(
    plot_asymmetry_levels,
    linear_performance_loss,
    "k-",
    marker="o",
    linewidth=2,
    label="Linear token-weighted voting"
)

ax.plot(
    plot_asymmetry_levels,
    quadratic_performance_loss,
    "k--",
    marker="s",
    linewidth=2,
    label="Quadratic voting"
)

ax.set_xlabel("Token Concentration", fontweight="bold", fontsize=12)
ax.set_ylabel("Performance Loss", fontweight="bold", fontsize=12)
ax.set_xticks(plot_asymmetry_levels)
ax.set_xticklabels([f"$\\rho={i}$" for i in plot_asymmetry_levels])

ax.legend(frameon=False, fontsize=11, prop=legend_properties)

plt.tight_layout()
plt.savefig(
    data_folder + "/Quadratic_vs_linear_asymmetry_mitigation.png",
    transparent=True,
    dpi=300
)
plt.show()
plt.clf()
```

```python
plot_asymmetry_levels = [1, 2, 3]

linear_performance_loss = [
    linear_dao_performance[0][-1] - linear_dao_performance[i][-1]
    for i in plot_asymmetry_levels
]

quadratic_performance_loss = [
    dao_performance[0][-1] - dao_performance[i][-1]
    for i in plot_asymmetry_levels
]

fig, ax = plt.subplots(figsize=(6, 4))

ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)

x = np.arange(len(plot_asymmetry_levels))
width = 0.35

ax.bar(
    x - width / 2,
    linear_performance_loss,
    width,
    edgecolor="black",
    facecolor="white",
    linewidth=1.5,
    label="Linear token-weighted voting"
)

ax.bar(
    x + width / 2,
    quadratic_performance_loss,
    width,
    edgecolor="black",
    facecolor="white",
    hatch="//",
    linewidth=1.5,
    label="Quadratic voting"
)

ax.set_xlabel(
    "Token Concentration",
    fontweight="bold",
    fontsize=12
)

ax.set_ylabel(
    "Performance Loss",
    fontweight="bold",
    fontsize=12
)

ax.set_xticks(x)
ax.set_xticklabels(
    [
        "High\n($\\rho=1$)",
        "Medium\n($\\rho=2$)",
        "Low\n($\\rho=3$)"
    ],
    fontsize=11
)

ax.legend(
    frameon=False,
    fontsize=11,
    prop=legend_properties,
    handlelength=3.0,   # default is ~2.0
)

plt.tight_layout()

plt.savefig(
    data_folder + "/Quadratic_vs_linear_asymmetry_mitigation_histogram.png",
    transparent=True,
    dpi=300
)

plt.show()
plt.clf()
```
