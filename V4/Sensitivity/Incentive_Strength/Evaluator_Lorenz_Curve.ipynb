```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
legend_properties = {'weight':'bold'}
```

```python
# Lorenz Curve
# Step 1: Randomly initialize a list of wealth values
np.random.seed(42)  # For reproducibility
wealth = np.random.randint(1, 100, 50)  # List of 50 random wealth values between 1 and 100

# Step 2: Sort the list of wealth values in ascending order
sorted_wealth = np.sort(wealth)

# Step 3: Calculate the cumulative share of wealth and the cumulative share of the population
cumulative_wealth = np.cumsum(sorted_wealth)
total_wealth = cumulative_wealth[-1]
cumulative_wealth_share = cumulative_wealth / total_wealth
cumulative_population_share = np.arange(1, len(wealth) + 1) / len(wealth)

# Step 4: Plot the Lorenz Curve
plt.figure(figsize=(10, 6))
plt.plot(cumulative_population_share, cumulative_wealth_share, marker='o', label='Lorenz Curve')
plt.plot([0, 1], [0, 1], linestyle='--', color='r', label='Line of Equality')
plt.title('Lorenz Curve')
plt.xlabel('Cumulative Share of Population')
plt.ylabel('Cumulative Share of Wealth')
plt.legend()
plt.grid(True)
plt.show()
```
