```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
legend_properties = {'weight':'bold'}
```

```python
data_folder = r"E:\data\dao-0310-23\V4_4\Asymmetry"
dao_performance_file = data_folder + r"\dao_performance"
consensus_performance_file = data_folder + r"\dao_consensus_performance"
dao_diversity_file = data_folder + r"\dao_diversity"
dao_variance_file = data_folder + r"\dao_variance"

dao_original_performance_file = data_folder + r"\dao_original_performance"

with open(dao_performance_file, 'rb') as infile:
    dao_performance = pickle.load(infile)
with open(dao_diversity_file, 'rb') as infile:
    dao_diversity = pickle.load(infile)
with open(dao_variance_file, 'rb') as infile:
    dao_variance = pickle.load(infile)

with open(consensus_performance_file, 'rb') as infile:
    consensus_performance = pickle.load(infile)

with open(dao_original_performance_file, 'rb') as infile:
    dao_original_performance = pickle.load(infile)
```

```python
# Define more colors
# NUS two colors
nus_blue = "#003D7C"
nus_orange = "#EF7C00"
# Nature three colors
nature_orange = "#F16C23"
nature_blue = "#2B6A99"
nature_green = "#1B7C3D"
# Morandi six colors
morandi_blue = "#046586"
morandi_green =  "#28A9A1"
morandi_yellow = "#C9A77C"
morandi_orange = "#F4A016"
morandi_pink = "#F6BBC6"
morandi_red = "#E71F19"
morandi_purple = "#B08BEB"
# Others
shallow_grey = "#D3D4D3"
deep_grey = "#A6ABB6"
```

```python
# Performance
x = range(len(dao_performance[0]))
fig, ax = plt.subplots()
ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)
plt.tight_layout()
plt.plot(x, dao_performance[0], "k-", label='$\\rho=0$')
plt.plot(x, dao_performance[1], "k:", label='$\\rho=1$')
plt.plot(x, dao_performance[2], "k--", label='$\\rho=2$')
plt.plot(x, dao_performance[3], "k-.", label='$\\rho=3$')
plt.xlabel('Time', fontweight='bold', fontsize=12)
plt.ylabel('Performance', fontweight='bold', fontsize=12)
# plt.xticks(x)
plt.legend(frameon=False, ncol=1, fontsize=12, prop=legend_properties)
plt.savefig(data_folder + r"\Asymmetry_performance.png", transparent=True, dpi=300)
plt.show()
plt.clf()
```

```python
# Diversity
x = range(len(dao_diversity[0]))
fig, ax = plt.subplots()
ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)
plt.tight_layout()
plt.plot(x, dao_diversity[0], "k-", label='$\\rho=0$')
plt.plot(x, dao_diversity[1], "k:", label='$\\rho=1$')
plt.plot(x, dao_diversity[2], "k--", label='$\\rho=2$')
plt.plot(x, dao_diversity[3], "k-.", label='$\\rho=3$')
plt.xlabel('Time', fontweight='bold', fontsize=12)
plt.ylabel('Diversity', fontweight='bold', fontsize=12)
# plt.xticks(x)
plt.legend(frameon=False, ncol=1, fontsize=12, prop=legend_properties)
plt.savefig(data_folder + r"\Asymmetry_diversity.png", transparent=True, dpi=300)
plt.show()
plt.clf()
```

```python
# Variance
x = range(len(dao_variance[0]))
fig, ax = plt.subplots()
ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)
num_markers = 10
# Calculate indices for marker positions
marker_positions = np.linspace(0, len(x) - 1, num=num_markers, dtype=int)
plt.plot(x, dao_variance[0], "-s", label='Equality', color=nature_orange, markevery=marker_positions)
plt.plot(x, dao_variance[1], "-v", label='$\\rho=1$', color=nature_green, markevery=marker_positions)
plt.plot(x, dao_variance[2], "-o", label='$\\rho=2$', color=nature_blue, markevery=marker_positions)
# plt.plot(x, dao_variance[3], "-", label='$\\rho=3$', color=nature_orange)
plt.xlabel('Time', fontweight='bold', fontsize=12)
plt.ylabel('Std. Dev. of Performance', fontweight='bold', fontsize=12)
# plt.xticks(x)
plt.legend(frameon=False, ncol=1, fontsize=12, prop=legend_properties)
plt.tight_layout()
plt.savefig(data_folder + r"\Asymmetry_sd.png", transparent=True, dpi=300, bbox_inches='tight')
plt.show()
plt.clf()
```
