```python
import matplotlib.pyplot as plt
import numpy as np

# Define the Morandi color palette
morandi_colors = ['#B08BEB', '#B5EAD7', '#F6D186', '#A9C9E3', '#F5B0CB']

# Generate some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.arctan(x)

# Create the figure and axis objects
fig, ax = plt.subplots()

# Plot the data as lines with Morandi colors
ax.plot(x, y1, color=morandi_colors[0])
# ax.plot(x, y2, color=morandi_colors[1])
# ax.plot(x, y3, color=morandi_colors[2])
# ax.plot(x, y4, color=morandi_colors[4])

# Set the x and y axis labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

# Set the axis limits
ax.set_xlim(0, 10)
ax.set_ylim(-1, 1)

# Save the figure to a file
fig.savefig('line_plot.png')
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure object and a set of subplots
fig, ax = plt.subplots()

# Define the x-axis range
ax.set_xlim([0, 1])

# Define the y-axis range
ax.set_ylim([0, 1])

# Create a rectangle patch
rect = patches.Rectangle((0.5, 0), 0.1, 1, alpha=0.5)

# Add the rectangle patch to the plot
ax.add_patch(rect)

# Show the plot
plt.show()
```
