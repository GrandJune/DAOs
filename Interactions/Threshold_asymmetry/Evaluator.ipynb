```python
asymmetry_list = [1, 2, 3]
data_folder = r"E:\data\dao-0310-23\V5_interaction\Threshold_asymmetry"
dao_performance_file_1 = data_folder + r"\dao_performance_across_threshold_1"
dao_consensus_file_1 = data_folder + r"\dao_consensus_performance_across_threshold_1"
dao_diversity_file_1 = data_folder + r"\dao_diversity_across_threshold_1"
dao_variance_file_1 = data_folder + r"\dao_variance_across_threshold_1"

dao_performance_file_2 = data_folder + r"\dao_performance_across_threshold_2"
dao_consensus_file_2 = data_folder + r"\dao_consensus_performance_across_threshold_2"
dao_diversity_file_2 = data_folder + r"\dao_diversity_across_threshold_2"
dao_variance_file_2 = data_folder + r"\dao_variance_across_threshold_2"

dao_performance_file_3 = data_folder + r"\dao_performance_across_threshold_3"
dao_consensus_file_3 = data_folder + r"\dao_consensus_performance_across_threshold_3"
dao_diversity_file_3 = data_folder + r"\dao_diversity_across_threshold_3"
dao_variance_file_3 = data_folder + r"\dao_variance_across_threshold_3"


with open(dao_performance_file_1, 'rb') as infile:
    dao_performance_1 = pickle.load(infile)
with open(dao_variance_file_1, 'rb') as infile:
    dao_variance_1 = pickle.load(infile)
with open(dao_diversity_file_1, 'rb') as infile:
    dao_diversity_1 = pickle.load(infile)
with open(dao_consensus_file_1, 'rb') as infile:
    dao_consensus_1 = pickle.load(infile)

with open(dao_performance_file_2, 'rb') as infile:
    dao_performance_2 = pickle.load(infile)
with open(dao_variance_file_2, 'rb') as infile:
    dao_variance_2 = pickle.load(infile)
with open(dao_diversity_file_2, 'rb') as infile:
    dao_diversity_2 = pickle.load(infile)
with open(dao_consensus_file_2, 'rb') as infile:
    dao_consensus_2 = pickle.load(infile)

with open(dao_performance_file_3, 'rb') as infile:
    dao_performance_3 = pickle.load(infile)
with open(dao_variance_file_3, 'rb') as infile:
    dao_variance_3 = pickle.load(infile)
with open(dao_diversity_file_3, 'rb') as infile:
    dao_diversity_3 = pickle.load(infile)
with open(dao_consensus_file_3, 'rb') as infile:
    dao_consensus_3 = pickle.load(infile)
    
dao_reference_file = r"E:\data\dao-0310-23\V4_4\Threshold\dao_performance_across_threshold"
dao_reference_variance_file = r"E:\data\dao-0310-23\V4_4\Threshold\dao_variance_across_threshold"
with open(dao_reference_file, 'rb') as infile:
    dao_reference = pickle.load(infile)
with open(dao_reference_variance_file, 'rb') as infile:
    dao_reference_variance = pickle.load(infile)
```

```python
# Define more colors
# NUS two colors
nus_blue = "#003D7C"
nus_orange = "#EF7C00"
# Nature three colors
nature_orange = "#F16C23"
nature_blue = "#2B6A99"
nature_green = "#1B7C3D"
# Morandi six colors
morandi_blue = "#046586"
morandi_green =  "#28A9A1"
morandi_yellow = "#C9A77C"
morandi_orange = "#F4A016"
morandi_pink = "#F6BBC6"
morandi_red = "#E71F19"
morandi_purple = "#B08BEB"
# Others
shallow_grey = "#D3D4D3"
deep_grey = "#A6ABB6"
```

```python
# Performance
x = np.arange(0.40, 0.71, 0.01)
fig, (ax1) = plt.subplots(1, 1)
ax1.plot(x, dao_reference, "-", label="Equality", color=deep_grey)
ax1.plot(x, dao_performance_1, "-", label="$\\rho=1$", color=nature_orange)  # most asymmetry
ax1.plot(x, dao_performance_2, "--", label="$\\rho=2$", color=nature_green)
ax1.plot(x, dao_performance_3, ":", label="$\\rho=3$", color=nature_blue)
plt.xlabel('Threshold', fontweight='bold', fontsize=12)
plt.ylabel('Performance', fontweight='bold', fontsize=12)
# plt.xticks(x)
handles, labels = ax1.get_legend_handles_labels()
handles = [h[0] if isinstance(h, container.ErrorbarContainer) else h for h in handles]
plt.legend(frameon=False, ncol=2, fontsize=12)
plt.savefig(data_folder + r"\Performance_across_threshold_asymmetry.png", transparent=True, dpi=300)
plt.show()
```

```python
# Variance
x = np.arange(0.40, 0.71, 0.01)
fig, (ax1) = plt.subplots(1, 1)
ax1.plot(x, dao_reference_variance, "-", label="Equality", color=deep_grey)
ax1.plot(x, dao_variance_1, "-", label="$\\rho=1$", color=nature_orange)  # most asymmetry
ax1.plot(x, dao_variance_2, "--", label="$\\rho=2$", color=nature_green)
ax1.plot(x, dao_variance_3, ":", label="$\\rho=3$", color=nature_blue)
plt.xlabel('Threshold', fontweight='bold', fontsize=12)
plt.ylabel('Variance', fontweight='bold', fontsize=12)
# plt.xticks(x)
handles, labels = ax1.get_legend_handles_labels()
handles = [h[0] if isinstance(h, container.ErrorbarContainer) else h for h in handles]
plt.legend(frameon=False, ncol=2, fontsize=12)
plt.savefig(data_folder + r"\Variance_across_threshold_asymmetry.png", transparent=True, dpi=300)
plt.show()
```
